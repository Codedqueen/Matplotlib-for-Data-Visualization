### Introduction to Matplotlib
##### Matplotlib : This is one of the most widely used and most popular  data visualization library in python.
##### It was created by John Hunter(1968 -2012) who was a neubiologist and was part of a research team that 
##### was working on analizing eletrogragpic signals(ECOG).The yeam was using a proprietry software for the
##### analysis but they had only one licence and where taking turns in using it.In order to overcome this 
##### limitation, John set out to replace the proparatory software with a Matlab  based region that could
##### be utilized by him and his team mates and could be extended by multiple investigators.As a result 
##### Matplotlib was initially developed as an EEG/ECOG Visualization tool and just like Matlab,Matplotlib
##### was equipped with a scripting interface for quick and easy generation of graphics represented by pyplot.
##### (Analogous to matlab scriping interface).

#### Matplotlib Architecture
##### This is composed of 3 main layers namely,
##### 1. The backend layer (figure canvas, Renderer,Event)
##### 2. The Artist layer(Artist): most of the heavy lifting happens here.It is usually the appropriate programming when 
##### writing a Web application server,U.I application or a script to be shared with other developers.
##### 3.Scripting layer (pyplot): This is the apropriate layer for everyday purposes.it is considered
##### as the widest scripting interface to simplify common task and for quick and easy generation of graphics and plots.

#### BACKEND LAYER
##### It has 3 built-in abstract interface classes
##### i. Figure canvas : matplotlib.darkend_bases.FigureCanvas.It encompasses the area the figure is drawn.
##### ii. Randerer :matplotlib.darkend_bases.Renderer.It knows how to draw on the FigureCanvas,
##### iii. Event: matplotlib.backend_bases.Event .It handles user input such as keyboard strokes and mouse clicks.

#### ARTIST
##### It is comprised  of one main object which is the Artist.it knows how to use the Rendererto draw the Canvas
##### Everything you see on a matplotlib figure such as title,lines, tick labels and images are all an individual Artist instances.
##### There are two types of Artist Object;
##### A. Primitive: Line2D, Rectangle,Circle and Text
##### B. Composite:Axis,Tick,Axes,Figure

##### The third level matplotlib object that contains or manages all the element in a given graphic is the Figure Artist and the most
##### composete Artist is the Axes, this is because this is where most of Api plotting matrix is defined , including matirx to
##### create and manipulate te Ticks, The Axis linemthe plot background.
##### It is important to note that each composite Artist may contain other composite artist as well as Primitive Artist.
##### or example, a Figure artist may cointain an Axis, Rectangle or Text Artist

### Puting the Artist layer to use
#### Lets try to generate a Histogram of 10,000 random numbers  using the Artist Layer

```python
from matplotlib.backends.backend_agg  import FigureCanvasAgg as FigureCanvas  #import FigureCanvas
from matplotlib.figure import Figure # import Figure Artist
fig = Figure()
Canvas = FigureCanvas(fig)

#Create 10000 random numbers using Numpy
import numpy as np
x = np.random.randn(10000)

ax = fig.add_subplot(111) #create an Axes Artist

ax.hist(x, 100) # Generate a histogram of the 10000 numbers

#Add a title to the figure and save it
ax.set_title('Normal Distribution with $\mu=0, \sigma=l$')
fig.savefig('matplotlib_histogram.png')
```

```python
### scipting Layer
##### The scripting layer was develpoed for scientist who are not professional programmers.
##### The artist layer is syntatically heavy,this is because it is meant for developers and not for individuals
##### whose goal is to perform quick experimentary analysis of some data.
##### Matplotlib Scripting layer is essentially the matplotlib.pyplot.
##### It comprises mainly of pyplot interface, a scriping interface that is lighter than the Artist layer.
##### It admits the process of defining a canvas and a figure Artist instance and connecting them.
```

```python
# How to use pyplot interface to generate a random number of 10000 as we did using the Artist layer to draw an histogram,
#using the 10000 random number we generated .

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
plt.hist(x,100)

#Add a title to the figure and save it
plt.title(r'Normal Distribution with $\mu=0, \sigma=l$')
plt.savefig('matplotlib_histogram.png')
plt.show()
```

# Pie Charts

##### pie chart is a type of graph in which a circle is divided into sectors that each represent a proportion
##### of the whole.
##### A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices to 
##### illustrate numerical proportion. 
#####  As a beginner is adviseable you look up the documentation on pie chart from matplotlib.pyplot.pie to
#####  enable you know the right code to use.

```python
fifa.head(5)
```

```python
fifa['Preferred Foot']
```

```python
# To look up a specific index we use loc
# because we are going to pass it through a pie chart we will need to use count( the count gives us an array
# To check it out  just put left or right 

left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
# left
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
# right
# To make a pie chart pass in a list of numbers else your work will be messed up, 
# note , our left and right contains a list of numbers in array
plt.pie([left, right])
labels = ['left', 'right']
colors = ['#abcdef', '#aabbcc']
# use pli.pie to plot a piechart as seen from the documentation
# use autopct='%.2f %% 'to input percentage in 2 decimal places, 2f , 
# or 1f or 3f reps 3 or one decimal places as the case may be.

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

# To put a title to our pie chart
plt.title('Foot Preference Of FIFA Players')
plt.show()
```

```python
# Pie chart 2, a more complicated pie chart
# we will look at the break down of the weight of the fifa players and plot it as a pie chart
# you can look at the weight using (fifa.Weight)
```

```python
# To reset the weight to integer and remove Ibs)
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]   
heavy = fifa[(fifa.Weight >= 200)].count()[0]   
                     
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200','over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight Distribution of FIFA Players (in lbs')

plt.pie(weights, labels=labels, colors=colors, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.savefig('Fifa_Players_figure.png', dpi=300)
plt.show()                     
```
